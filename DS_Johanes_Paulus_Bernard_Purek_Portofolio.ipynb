```python
def remove_outliers_iqr(data, column_name):

  Q1 = data[column_name].quantile(.25)
  Q3 = data[column_name].quantile(.75)

  IQR = Q3 - Q1

  maximum = Q3 + (1.5 * IQR)
  minimum = Q1 - (1.5 * IQR)

  outlier_filter = (data[column_name] < minimum) | (data[column_name] > maximum)
  data = data[~outlier_filter]
  return data
```
